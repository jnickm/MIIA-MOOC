### 1. Modo comando.

Este modo se activa presionando la tecla `Esc` y/o dando click sobre la celda la cual se quiere interactua. Sabremos que estamos en el modo comando cuando la izquierda de la celda tenga una barra de color azul.

#### Atajos para correr codigo o representar texto.

 Empecemos esta sección con los que probablemente serán (o son) su mejor opción para evaluar el contenido que se ha escrito en la celda: 
 
   * `Control + Enter :`  evalúa el contenido de la celda y permanece en ella.

   * `Shift + Enter :`   evalúa el contenido de la celda y se mueve hacia la siguiente, creando una nueva si no hay otra celda a la cual se pueda mover.


#### Movimiento sobre el Notebook.


* $\downarrow$: Se desplaza hacia la celda inferior.
* $\uparrow$: Se deplaza hacia la celda superior.
* `A`: Añadir celda sobre la celda actual.
* `B`: Añadir celda bajo la celda actual. 
* `D + D`: Elimina la celda actual.
* `Z`    : Deshace la eliminación de una celda.
* `M`    : Convierte la celda actual en una celda Markdown.
* `C`     : Convierte la celda actual en una celda de codigo.

## Celdas.

Los Notebooks de Jupyter tienen dos tipos de celdas diferentes: celdas Markdown o celdas de texto y celdas de código.


### 1. Celdas Markdown.

Este tipo de celdas son las encargadas de almacenar el texto del documento, crear tablas, insertar imágenes, entre otras funciones. Es generalmente utilizado para la descripción, o explicación del código o resultados que se presentan en su(s) casilla(s) inferior(es).

Este tipo de celdas requieren de diferentes recursos para dar formato al texto que se va a presentar. Estos recursos son interpretados por el notebook para dar un resultado final, como puede ser **agregar texto en negrilla**, creación de formulas matemáticas: $\sum\limits_{i = 1}^{N}\kappa = \frac{n(n+1)}{2}$, o definición de títulos, sub-títulos, entre otras funciones que se explicarán a continuación.


#### 1.1  Encabezados:

# Títulos.
`#`    : Títulos
## Subtítulos.
`##`   : Subtítulos
### Apartados.
`###`  : Apartados
#### Subapartados.
`####` : Subapartados
##### sub-subapartados.
`#####` : Sub-subapartados


####  1.2 Tipografía.

Es posible escribir nuestros utilizar texturas en nuestras oraciones, ya sea palabras en **negrilla**, *cursiva*, ~~tachado~~ y **_negrilla y cursiva a la vez_**.

En la tabla a continuación se podrán ver los caracteres que debemos emplear para obtener este tipo de texturas.


|Recurso            |Resultado        |
|-------------------|-----------------|
|`**negrilla`       |**negrilla**     |
|`__negrilla__`     | __negrilla__    |
|`*cursiva*`        |*cursiva*        |
|`_cursiva_`        |_cursiva_        |
|`~~tachado~~`      |~~tachado~~      |
|`**_combinación_**`|**_combinacion_**|




De igual forma se puede agregar textura a una oración agregando un link en ella, esto se logra de utilizando la siguiente convención:

`[Aqui va la oración](http://https://uniandes.edu.co)`  

[Aqui va la oración](http://https://uniandes.edu.co)

####   1.3 Formulas matematicas.

Las ecuaciones en Markdown estan soportadas por $\LaTeX$. Si se desea utilizar una en el documento basta con escribir la formula en $\LaTeX$  entre los signos de dinero
; `$Aquí la ecuacion$`. 

Por ejemplo de la siguiente convencion `$\sum\limits_{i = 1}^30\frac{\theta}{n^3}={\pi^2}$`, se obtiene esta ecuacion: 
          $\sum\limits_{i = 1}^30\frac{\theta}{n^3}={\pi^2}$   

Si se requiere, en el siguiente [link](http://metodos.fam.cie.uva.es/~latex/apuntes/apuntes3.pdf) se pueden encontrar diferentes convenciones para la creacion de formulas matematicas escritas en $\LaTeX$. En este se puede ver la forma de escribir letras griegas, integrales, sumatorias, entre otros caracteres empleados cuando queremos escribir formulas en Markdown.
